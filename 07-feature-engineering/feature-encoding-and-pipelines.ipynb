ONE-HOT Encoding by OneHotEncoder

```python
X = df.drop('price', axis=1)
Y = df['price']
X
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output=False, drop='first')
X_transformed = ohe.fit_transform(X[['mainroad', 'guestroom', 'basement', 'airconditioning', 'furnishingstatus']])
X_transformed
```

```python
X_new = np.hstack((X_transformed,X[['area', 'bedrooms', 'bathrooms', 'parking']]))
X_new.shape
```

Column Transform

```python
from sklearn.compose import ColumnTransformer
cate_cols = X.select_dtypes(include='object').columns.tolist()
cate_cols
```

```python
preprocessor = ColumnTransformer(transformers=[('cat',OneHotEncoder(drop='first'), cate_cols)], remainder='passthrough')
```

```python
X_new_new = preprocessor.fit_transform(X)
X_new_new.shape
```

```python
preprocessor
```

PIPELINE

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
```

```python
pipeline = Pipeline(steps = [
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])
pipeline
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, random_state=42, test_size=0.2)
x_train.shape
```

```python
pipeline.fit(x_train,y_train)
```

```python
pipeline.score(x_test,y_test)
```
