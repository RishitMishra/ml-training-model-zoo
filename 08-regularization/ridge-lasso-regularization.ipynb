```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge,Lasso
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.datasets import fetch_california_housing
```

```python
data = fetch_california_housing()
```

```python
print(data.DESCR)
```

```python
# Create DataFrame
df = pd.DataFrame(data.data, columns = data.feature_names)
df['Target'] = data.target
# Split features and target
X = df.drop('Target', axis = 1)
Y = df['Target']
```

```python
df.head()
```

```python
df.shape
```

```python
# train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 42)
```

# Using Standard Scaler

```python
# Feature Scaling by Standard Scaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
ridge = Ridge(alpha = 1.0)
lasso = Lasso(alpha = 0.1)
lr = LinearRegression()
```

```python
ridge.fit(X_train_sc,Y_train)
```

```python
lasso.fit(X_train_sc,Y_train)
```

```python
lr.fit(X_train_sc,Y_train)
```

```python
ridge_pred = ridge.predict(X_test_sc)
```

```python
lasso_pred = lasso.predict(X_test_sc)
```

```python
lr_pred = lr.predict(X_test_sc)
```

```python
ridge_pred
```

```python
lasso_pred
```

```python
lr_pred
```

```python
print("Ridge R²:", r2_score(Y_test, ridge_pred))
print("Ridge MSE:", mean_squared_error(Y_test, ridge_pred))
```

```python
print("Lasso R²:", r2_score(Y_test, lasso_pred))
print("Lasso MSE:", mean_squared_error(Y_test, lasso_pred))
```

```python
print("LR R²:", r2_score(Y_test, lr_pred))
print("LR MSE:", mean_squared_error(Y_test, lr_pred))
```

# Using Polynomial

```python
# Feature Scaling by Polynomial Scaler
poly = PolynomialFeatures(degree = 2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train_poly,Y_train)
```

```python
lr_pred2 = lr.predict(X_test_poly)
```

```python
lr_pred2
```

```python
print("LR R²:", r2_score(Y_test, lr_pred2))
print("LR MSE:", mean_squared_error(Y_test, lr_pred2))
```
