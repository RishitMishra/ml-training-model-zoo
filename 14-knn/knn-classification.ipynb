# K-Nearest Neighbors Algorithm :- K-Nearest Neighbors (KNN) is a non-parametric, instance-based learning algorithm used for classification and regression.

```python
import pandas as pd 
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df = df[['survived','pclass','sex','age','fare','embarked']]
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
age_imputer = SimpleImputer(strategy='mean')
df['age'] = age_imputer.fit_transform(df[['age']])
embarked_imputer = SimpleImputer(strategy='most_frequent')
df['embarked'] = embarked_imputer.fit_transform(df[['embarked']])[:,0]
```

```python
df.dropna(inplace = True)
df.isnull().sum()
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
le = LabelEncoder()
df['sex'] = le.fit_transform(df['sex'])
df['embarked'] = le.fit_transform(df['embarked'])
```

```python
X = df.drop(columns = ['survived'])
Y = df['survived']
```

```python
sc = StandardScaler()
X_scaled = sc.fit_transform(X)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X_scaled,Y,test_size = 0.2,random_state = 42)
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors = 7)
clf.fit(X_train,Y_train)
```

```python
clf.score(X_test,Y_test)
```

```python
Y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,Y_pred))
```

```python
df = pd.read_csv('Salary_Data.csv')
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df['Education Level'].unique()
```

```python
df['Education Level'] = df['Education Level'].map({"Bachelor's Degree":"Bachelor's","Master's Degree":"Master's", 'phD':'PhD', "Bachelor's":"Bachelor's", "Master's":"Master's", 'PhD':'PhD', 'High School':'High School'})
```

```python
freq = df['Job Title'].value_counts().to_dict()
df['Job Title'] = df['Job Title'].map(freq)
df
```

```python
numerical = ['Age', 'Years of Experience']
```

```python
X = df.drop('Salary', axis=1)
Y = df['Salary']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, random_state=42, test_size=0.2)
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsRegressor
```

```python
preprocessor = ColumnTransformer([
    ('scaler', StandardScaler(), numerical),
    ('onehot', OneHotEncoder(drop='first'), ['Gender',]),
    ('ordinal', OrdinalEncoder(),['Education Level',])
])
```

```python
model = KNeighborsRegressor(n_neighbors=10)
```

```python
regressor = Pipeline([
    ('preprocessing', preprocessor),
    ('model', model)
])
```

```python
regressor.fit(x_train,y_train)
```

```python
regressor.score(x_test,y_test)
```
