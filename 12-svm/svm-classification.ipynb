```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
from sklearn.datasets import load_breast_cancer
df = load_breast_cancer()
```

```python
df.DESCR
```

```python
x = df.data
y = df.target
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
train_sc = scaler.fit_transform(x_train)
test_sc = scaler.transform(x_test)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='rbf', C=1.0, gamma='scale')
```

```python
model.fit(train_sc, y_train)
```

```python
pred = model.predict(test_sc)
```

```python
from sklearn.metrics import classification_report, accuracy_score

print(accuracy_score(y_test, pred))
```

```python
print(classification_report(y_test, pred))
```

```python
df1 = pd.read_csv('Salary_Data.csv')
df1
```

```python
df1.isnull().sum()
```

```python
df1.dropna(inplace=True)
```

```python
df1.isnull().sum()
```

```python
df1['Education Level'].unique()
```

```python
df1['Education Level'] = df1['Education Level'].map({"Bachelor's":1, "Master's":2, 'PhD':3, "Bachelor's Degree":1,
       "Master's Degree":2, 'High School':0, 'phD':3})
```

```python
df1
```

```python
df1['Gender'].unique()
```

```python
df1['Gender'] = df1['Gender'].map({'Male':1, 'Female':0, 'Other':2})
```

```python
X = df1.drop(columns=['Salary'], axis=1)
Y = df1['Salary']
```

```python
numerical = ['Age', 'Years of Experience']
category = ['Gender', 'Job Title', 'Education Level']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, random_state=42, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

scaler = StandardScaler()
onehot = OneHotEncoder(drop='first', handle_unknown='ignore')
```

```python
from sklearn.pipeline import Pipeline

preprocessor = ColumnTransformer([
    ('num', scaler, numerical),
    ('cat', onehot, category)
])
```

```python
from sklearn.svm import SVR
model = SVR(kernel='rbf',C=10000.0)
```

```python
clf = Pipeline([
    ('preprocessing', preprocessor),
    ('model', model)
])
```

```python
clf.fit(x_train, y_train)
```

```python
pred = clf.predict(x_test)
3
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_percentage_error

print(mean_squared_error(y_test, pred))
```

```python
print(r2_score(y_test, pred))
```

```python
print(mean_absolute_percentage_error(y_test, pred))
```
